# Extracting and displaying the text of the review at index 50 for analysis

```python
example = df['Text'][50]
print(example)
```

```python
tokens = nltk.word_tokenize(example)
tokens[:10]
```

```python
tagged = nltk.pos_tag(tokens)
tagged[:10]
```

```python
entities = nltk.chunk.ne_chunk(tagged)
entities.pprint()
```

# Sentiment Analysis with SentimentIntensityAnalyzer


```python
from nltk.sentiment import SentimentIntensityAnalyzer
sia = SentimentIntensityAnalyzer()
sia.polarity_scores('I am so happy!')
```

```python
sia.polarity_scores(example)
```

```python
# Run the polarity score on the entire dataset
res = {}
for i, row in df.iterrows():
    text = row['Text']
    myid = row['Id']
    res[myid] = sia.polarity_scores(text)
```

```python
res
```

# Creating a DataFrame for VADER Sentiment Scores


```python
vaders = pd.DataFrame(res).T
vaders = vaders.reset_index().rename(columns={'index': 'Id'})
vaders = vaders.merge(df, how='left')
vaders.head()
```

# Visualizing Compound Sentiment Scores by Amazon Star Rating


```python
ax = sns.barplot(data=vaders, x='Score', y='compound')
ax.set_title('Compund Score by Amazon Star Review')
plt.show()
```

# Analyzing Positive, Neutral, and Negative Sentiment Scores by Amazon Star Rating


```python
fig, axs = plt.subplots(1, 3, figsize=(12, 3))
sns.barplot(data=vaders, x='Score', y='pos', ax=axs[0])
sns.barplot(data=vaders, x='Score', y='neu', ax=axs[1])
sns.barplot(data=vaders, x='Score', y='neg', ax=axs[2])
axs[0].set_title('Positive')
axs[1].set_title('Neutral')
axs[2].set_title('Negative')
plt.tight_layout()
plt.show()
```
